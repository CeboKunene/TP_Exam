```python
raws = 5
column_Name = 'Product_Name'
df_column = df[column_Name]
df_raw = df_column.head(raws)

column = 'totalAmount'
Name = df[column]
dataH = Name.head(raws)


plt.pie(dataH, labels=df_raw,autopct='%1.1f%%',startangle=140)
plt.title('Product_Name AND totalAmount')
plt.show()
```

```python

sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot( x=df_raw, y=dataH, data=df)
plt.xlabel('Product_Name')
plt.ylabel('totalAmount')
plt.show()
```

```python
# Calculate mean, median, and standard deviation only for numeric columns
numeric_data = df.select_dtypes(include=['float64', 'int64'])

print("\nMean of numeric columns:\n", numeric_data.mean())
print("\nMedian of numeric columns:\n", numeric_data.median())
print("\nStandard deviation of numeric columns:\n", numeric_data.std())
```

```python



plt.figure(figsize=(10, 6))
sns.barplot(x='Product_Name', y='totalAmount', data=df)
plt.title('Population Amount by totalAmount')
plt.xlabel('Product_Name')
plt.ylabel('totalAmount')
plt.xticks(rotation=45)
plt.show()
```
